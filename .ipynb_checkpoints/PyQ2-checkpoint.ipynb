```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('Book3.xlsx', sheet_name='Sheet2')
df.to_csv('file2.csv',index=False)
df.head()
```

```python
# Drop rows with missing values
df = df.dropna()
```

```python
# 1. Correlation Matrix
corr = df.corr(numeric_only=True)
plt.figure(figsize=(10, 8))
sns.heatmap(corr[['Sales']].sort_values(by='Sales', ascending=False), annot=True, cmap='coolwarm')
plt.title("Correlation with Sales")
plt.show()
```

```python
# 2. Top 5 Positively Correlated Features with Sales
top_corr = corr['Sales'].sort_values(ascending=False).drop('Sales').head(5)
print("Top 5 positively correlated features with Sales:\n", top_corr)
```

```python
# 3. Average Sales by Holiday
print("\nAverage Sales by Holiday:")
print(df.groupby('Holiday Dummy')['Sales'].mean())
```

```python
# 4. Discount impact on Sales
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='Discount1', y='Sales')
plt.title("Sales vs Discount1")
plt.show()
```

```python
# 5. Total SKU vs Sales
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='Total SKU', y='Sales')
plt.title("Sales vs Total SKU")
plt.show()
```

```python
# 6. Paid Search Impressions vs Sales
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='Paid Search Impressions', y='Sales')
plt.title("Sales vs Paid Search Impressions")
plt.show()
```

```python
# 7. Organic Search Impressions vs Sales
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='Organic Search Impressions', y='Sales')
plt.title("Sales vs Organic Search Impressions")
plt.show()
```

```python
# 8. Average Price vs Sales
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='Average Price', y='Sales')
plt.title("Sales vs Average Price")
plt.xlabel("Average Price")
plt.ylabel("Sales")
plt.show()
```

```python
# 9. Gasoline Price vs Sales
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='Gasoline Price', y='Sales')
plt.title("Sales vs Gasoline Price")
plt.xlabel("Gasoline Price")
plt.ylabel("Sales")
plt.show()
```

```python
# 10. Email Clicks vs Sales
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='Email Clicks', y='Sales')
plt.title("Sales vs Email Clicks")
plt.xlabel("Email Clicks")
plt.ylabel("Sales")
plt.show()
```
